### K-Means (distance between points)

Load libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from sklearn.preprocessing import Normalizer
from sklearn.metrics.cluster import silhouette_score
```

Load data

```python
matrix=np.array(np.loadtxt('matrix.txt'))
```

Normalization and plot

```python
normalizer_scaler=Normalizer()
normalizer_scaler.fit(matrix)
matrix_normalizer=normalizer_scaler.transform(matrix)
plt.title(label="Нормування даних методом Normalizer")
plt.grid()
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(matrix_normalizer[:,0],matrix_normalizer[:,1],label="Normalizer")
plt.show()
```

Using K-Means and plot results

```python
X=matrix
kmeans=KMeans(n_clusters=2)
kmeans.fit(X)
y_pred=kmeans.predict(X)
plt.scatter(X[:,0],X[:,1],c=y_pred,s=60)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='^',s=100,linewidth=2)
plt.xlabel("Перша ознака")
plt.ylabel("Друга ознака")
plt.title("Кластеризація методом k-середніх")
plt.show()
#Task 4
print("Аналіз якості кластеризації за допомогою обчислення силуетної метрики K-means:\t",silhouette_score(X,kmeans.labels_,metric='euclidean'))
```

Using K-Means with normalized data and plot results

```python
kmeans=KMeans(n_clusters=2)
kmeans.fit(matrix_normalizer)
y_pred=kmeans.predict(matrix_normalizer)
plt.scatter(matrix_normalizer[:,0],matrix_normalizer[:,1],c=y_pred,s=60)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='^',s=100,linewidth=2)
plt.xlabel("Перша ознака")
plt.ylabel("Второй признак")
plt.title("Кластеризація методом k-середніх нормалізованих даних")
plt.show()
#Task 4
print("Аналіз якості кластеризації за допомогою обчислення силуетної метрики K-means нормалізованих даних:\t",silhouette_score(matrix_normalizer,kmeans.labels_,metric='euclidean'))
```

### Data-Mining-Algorithmus zur Clusteranalyse. (distance between nearest points)

Using DBSCAN and plot results

```python
Y=np.array(np.loadtxt('matrix.txt'))
dbscan=DBSCAN(eps=7,min_samples=3)
clusters=dbscan.fit_predict(Y)
plt.scatter(Y[:,0],Y[:,1],c=clusters,s=60)
plt.xlabel("Перша ознака")
plt.ylabel("Друга ознака")
plt.title("Кластеризація методом DBSCAN")
plt.show()
print("Аналіз якості кластеризації за допомогою обчислення силуетної метрики DBSCAN:\t",silhouette_score(Y,dbscan.labels_))
```

Using DBSCAN with normalized data and plot results

```python
Y=matrix_normalizer
dbscan=DBSCAN(eps=0.05,min_samples=3)
clusters=dbscan.fit_predict(Y)
plt.scatter(Y[:,0],Y[:,1],c=clusters,s=60)
plt.xlabel("Перша ознака")
plt.ylabel("Второй признак")
plt.title("Кластеризація методом DBSCAN Нормовані дані")
plt.show()

#TODO dbscan.labels
print("Аналіз якості кластеризації за допомогою обчислення силуетної метрики DBSCAN нормованих даних:\t",silhouette_score(Y,dbscan.labels_))
```
